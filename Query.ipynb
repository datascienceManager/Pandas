```python
import pandas as pd
import numpy as np
```

```python
TechData = { 'Course':['Spark','Pyspark','Hadoop','Python','Pandas'],
             'Fee': [2000,5000,4000,1200,5600],
             'Duration':['30days','15days','14days',None,np.nan],
             }

TechData = pd.DataFrame(TechData)

print(TechData)
```

```python
df2 = TechData.query("Course == 'Pyspark'")

df2
```

```python
# Use a variable in the expression, use @ character

value = "Pandas"

df3 = TechData.query("Course == @value")

df3
```

```python
# Updating the existing DataFrame use inplace=True

data = TechData

data.query("Course == 'Pandas'", inplace= True)

data
```

```python
# column value not equals then use != operator

df5 = TechData.query("Course != 'Pyspark'")

df5
```

```python
# Select Rows Based on List of Column Values

df7=TechData.query("Course in ('Spark','Pandas')")

df7
```

```python
# Query Rows by list of values

values = ('Spark','Pandas')

df8 = TechData.query("Course in @values")

df8
```

```python
# Not in list 

df9 = TechData.query("Course not in @values")

df9
```

```python
# Query with Multiple Conditions

df10 = TechData.query("`Fee`>= 2000 and `Fee` <= 5000")

df10
```

```python
# create sample dataframe
df = pd.DataFrame(np.random.randint(20, 100, size=(100,4)),
                  columns=['A', 'B', 'C', 'D'])

df
```

pd.eval() and pd.query() come in handy to increase the computational efficiency

pd.eval() for Compound Expression

pd.eval() allows you to compute compound expression more efficiently

```python
df1 = df.eval('A >=50 and C <= 40')

dfdata = df[df1] 

dfdata
```

```python
df.eval('E = (A+B)',inplace= True)

print(df)
```

```python
df.eval('J = (A + A.mean())',inplace = True)

df
```
